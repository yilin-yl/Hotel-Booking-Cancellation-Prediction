# Data Cleaning and Preprocessing

### 1. Missing Value

```python
df.isnull().sum()
```

```python
#missing rate
total_data = df.shape[0]
df.isnull().sum()/total_data
```

Missing value part: Children, Country, Agent, Company

(1)children: we replace the 4 null values of "children" with 0

(2)country: assign "PRT" to the missing country

(3)agent: Impute with 'undefined' for 'agent'

(4)Company: 94% of this column is missing, so we drop it

```python
#Missing value part: Children, Country, Agent, Company 
df['children'] = df['children'].replace(np.nan, 0)
df['children'] = df['children'].astype(int)
# assign "PRT" to the missing country
country_na = 'PRT'
df['country'] = df['country'].replace(np.nan, country_na)
#Impute with '0' for 'agent'
df['agent'] = df['agent'].replace(np.nan, 0)
# drop 'company'
df = df.drop(['company'], axis = 1)
```

#### 3.2 Numerical variables

```python
def drop_outlier(name):
    values = df[name]
    sorted(values)
    q1,q3 = values.quantile([0.25,0.75])
    mini = q1 - (3*(q3-q1))
    maxi = q3 + (3*(q3-q1))
    index_names = df[(df[name]> maxi) | (df[name]< mini)].index
    #print(name + ' ' + str(maxi) + ' ' + str(mini))
    print('drop ' + str(len(index_names)) + ' rows from ' + name)
    df.drop(index_names, inplace = True)

list_ = ['adr','lead_time','stays_in_week_nights','stays_in_weekend_nights', 'total_of_special_requests']
```

```python
for i in list_:
    drop_outlier(i)
```

### 6. Feature Engineering

```python
# month
import calendar
month = []
for i in df.arrival_date_month:
    mon = list(calendar.month_name).index(i)
    month.append(mon)
df.insert(4,"arrival_month",month)
```

```python
# add a feature of room change
df['change_in_room'] = df.apply(lambda x: 1 if x['assigned_room_type'] == x['reserved_room_type'] else 0, axis = 1)
# add babies and children togather, further will drop babies and children
df['total_children'] = df.apply(lambda x: x['babies']+x['children'],axis=1)
# add total nights:
df['total_nights'] = df.apply(lambda x: x['stays_in_weekend_nights'] + x['stays_in_week_nights'], axis = 1)
df['weekend_weekday_ratio'] = df.apply(lambda x: 1 if x['stays_in_week_nights']== 0 else x['stays_in_weekend_nights'] / x['stays_in_week_nights'], axis = 1)
# group country: 
df['country_group'] = df.apply(lambda x: 'group1' if (x['country']== 'GBR' or x['country']=='FRA' or x['country']=='ESP' or x['country'] =='DEU') else x['country'],axis =1)
df['country_group'] = df.apply(lambda x: 'group2' if (x['country_group'] != 'PRT' and x['country_group'] != 'group1') else x['country_group'] ,axis =1)
df['country_group'] = df.apply(lambda x: 'PRT' if x['country'] == 'PRT' else x['country_group'] ,axis =1)
# group agent
df['agent_group'] = df.apply(lambda x:'group1' if (x['agent'] == 9 or x['agent'] == 0 or x['agent']==240) else 'group2', axis = 1)
```

```python
# drop country
df = df.drop(['country'], axis = 1)

#drop agent
df = df.drop(['agent'], axis = 1)

# children + babies = total_children 
df = df.drop(['babies'], axis = 1)
df = df.drop(['children'], axis = 1)

# arrival_date_week_number & arrival_date_year & arrival_date_month are correlated, over 0.5
#df = df.drop(['arrival_date_week_number'], axis = 1)
df = df.drop(['arrival_date_year'], axis = 1)
df = df.drop(['arrival_date_month'], axis = 1)
df = df.drop(['arrival_date_day_of_month'], axis = 1)
df = df.drop(['arrival_month'], axis = 1)

#stays in week nights and stays in weekend nights, correlation=0.38
df = df.drop(['stays_in_week_nights'], axis = 1)
df = df.drop(['stays_in_weekend_nights'], axis = 1)

# market_segment contains information of distribution_channel, so drop distribution_channel
df = df.drop(['distribution_channel'], axis = 1)

# room type:
df = df.drop(['reserved_room_type'], axis = 1)

# similar to dependent variable, not for our modeling
df = df.drop(['reservation_status'], axis = 1)
df = df.drop(['reservation_status_date'], axis = 1)
#df = df.drop(['days_in_waiting_list'], axis = 1)

# have correlated information in logic, previous booking is canceled previous booking not canceled & is repeated guest
df = df.drop(['is_repeated_guest'], axis = 1)
```

## Split Dataset

```python
#modelling
from sklearn.metrics import confusion_matrix
np.random.seed(0)
from sklearn.model_selection import train_test_split

#cross validation
#k-fold
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_city, y_city, test_size = 0.2)
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_resort, y_resort, test_size = 0.2)

Xc_train, Xc_val, yc_train, yc_val = train_test_split(Xc_train, yc_train, test_size = 0.2)
Xr_train, Xr_val, yr_train, yr_val = train_test_split(Xr_train, yr_train, test_size = 0.2)
```

```python
adr_r = df_resort['adr'].mean()
adr_c = df_city['adr'].mean()

def profit_(true, pred, adr, profit_rate):
    oper_rate = 1 - profit_rate
    tn, fp, fn, tp = confusion_matrix(pred, true).ravel()
    rooms = tn+fp+tp+fn
    profit_per_room = adr * profit_rate
    profit = tp*profit_per_room + fp*(profit_per_room*2 - adr) + fn*(-adr*oper_rate) + tn*(profit_per_room)
    print('profit per room is ' + str(profit/rooms))
    return profit/rooms
def c_m_analysis(true,pred):
    tn, fp, fn, tp = confusion_matrix(pred, true).ravel()
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    fpr = fp/(fp+tn)
    f_score = 2*precision*recall/(precision+recall)
    #
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    print("Precision:\t\t\t%1.2f identified as cancellations are cancellations"%(precision))
    print("Recall/TPR:\t\t\t%1.2f proportion of actual cancellations identified"%(recall))
    print("False Positive Rate:\t\t%1.2f proportion of check-ins identified as cancellations"%fpr)
    print("f-score:\t\t\t%1.2f tradeoff between precision and recall"%(f_score))
    print("Accuracy:\t\t\t%1.2f how well the model has classified"%(accuracy))
```

## 1. Logistic Regression

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score


#resort
model_lr_r = LogisticRegression(solver = 'liblinear').fit(Xr_train, yr_train)
pred_lr_r = model_lr_r.predict(Xr_val)
print("R-Square of Resort Hotel",r2_score(yr_val,pred_lr_r))
print('confusion matrix of Resort Hotel: \n', confusion_matrix(pred_lr_r,np.array(yr_val)))
c_m_analysis(yr_val,pred_lr_r)
print("Validation Accuray of Resort Hotel",model_lr_r.score(Xr_val, yr_val))
print("Reost hotel's per room profit: ", profit_(yr_val,pred_lr_r,adr_r,0.5))


print('\n')
#city
model_lr_c = LogisticRegression(solver = 'liblinear').fit(Xc_train, yc_train)
pred_lr_c = model_lr_c.predict(Xc_val)
print("R-Square of City Hotel",r2_score(yc_val,pred_lr_c))
print('confusion matrix of City Hotel: \n', confusion_matrix(pred_lr_c,np.array(yc_val)))
print("Validation Accuray of city Hotel",model_lr_c.score(Xc_val, yc_val))
c_m_analysis(yc_val,pred_lr_c)
print("City hotel's per room profit: ", profit_(yc_val,pred_lr_c,adr_c,0.6))
```
